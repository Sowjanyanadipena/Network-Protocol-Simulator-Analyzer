```python
#  Network Protocol Simulation: Stop-and-Wait ARQ

import time
import random
import matplotlib.pyplot as plt

num_packets = 20
drop_rate = 0.2
delays = []
success = []

for i in range(num_packets):
    send_time = time.time()
    if random.random() < drop_rate:
        print(f"Packet {i+1} dropped.")
        delays.append(None)
        success.append(0)
        time.sleep(0.5)
    else:
        delay = round(random.uniform(0.1, 0.4), 3)
        time.sleep(delay)
        ack_time = time.time()
        rtt = round(ack_time - send_time, 3)
        delays.append(rtt)
        success.append(1)
        print(f"Packet {i+1} sent successfully. RTT: {rtt}s")

# Plot RTT
plt.figure(figsize=(10, 5))
plt.plot([i+1 for i in range(num_packets)], [d if d else 0 for d in delays], marker='o', label='RTT')
plt.title("Packet RTT and Loss")
plt.xlabel("Packet #")
plt.ylabel("RTT (s)")
plt.grid(True)
plt.show()
```
